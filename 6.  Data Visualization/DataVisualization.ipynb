### 1. Create a scatter plot using Matplotlib to visualise the relationship between two arrays, x and y for the given data.

### 1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic dataset.

````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Generating synthetic data (same as before)
np.random.seed(0)
temperature = np.random.uniform(20, 40, 100)
ice_cream_sales = 2 * temperature + np.random.normal(0, 5, 100)

# Creating the scatter plot
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 7))

# Scatter plot with labels
sns.scatterplot(x=temperature, y=ice_cream_sales, color="skyblue", s=100, alpha=0.8, label="Data Points")

# Adding a trendline (Corrected)
slope, intercept, r_value, p_value, std_err = stats.linregress(temperature, ice_cream_sales)
plt.plot(temperature, slope * temperature + intercept, color="black", label=f"Trendline (R={r_value**2:.2f})")

# Annotations for key points (Corrected)
max_sales_index = np.argmax(ice_cream_sales)
min_sales_index = np.argmin(ice_cream_sales)
plt.annotate("Highest Sales", xy=(temperature[max_sales_index], ice_cream_sales[max_sales_index]), xytext=(temperature[max_sales_index]+1, ice_cream_sales[max_sales_index]-5),
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=-0.2"), fontsize=10)
plt.annotate("Lowest Sales", xy=(temperature[min_sales_index], ice_cream_sales[min_sales_index]), xytext=(temperature[min_sales_index]-2, ice_cream_sales[min_sales_index]+5),
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=0.2"), fontsize=10)

# Other plot elements
plt.title("Relationship between Temperature and Ice Cream Sales\n", fontsize=16)
plt.xlabel("Temperature (C)", fontsize=14)
plt.ylabel("Ice Cream Sales (Units)", fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, linestyle="--", alpha=0.5)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Displaying the plot
plt.tight_layout()
plt.show()
````

### 4. Create a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.

````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(0)

# Generate synthetic data
n = 200
categories = ['North', 'South', 'East', 'West']
category = np.random.choice(categories, size=n)
sales = np.random.normal(loc=100, scale=20, size=n)

# Create a DataFrame
data = pd.DataFrame({
    'Region': category,
    'Sales': sales
})

# Calculate overall mean
overall_mean = data['Sales'].mean()

# Create a box plot with a color palette and hue
plt.figure(figsize=(14, 8))
ax = sns.boxplot(data=data, x='Region', y='Sales', hue='Region', palette='Set2', legend=False)

# Add annotations for median values
for i, region in enumerate(data['Region'].unique()):
    median_value = data[data['Region'] == region]['Sales'].median()
    ax.text(i, median_value + 10, f'Median: {median_value:.2f}', 
            horizontalalignment='center', size='medium', color='black', weight='bold', ha='center')

# Add mean markers for each category
mean_values = data.groupby('Region')['Sales'].mean()
for i, region in enumerate(data['Region'].unique()):
    ax.plot(i, mean_values[region], 'o', color='red', label='Mean' if i == 0 else "", markersize=8, markeredgecolor='black')

# Add overall mean line
ax.axhline(overall_mean, color='purple', linestyle='--', linewidth=2, label='Overall Mean')

# Customize plot aesthetics
plt.title('Distribution of Sales Across Different Regions', fontsize=18, fontweight='bold')
plt.xlabel('Region', fontsize=15)
plt.ylabel('Sales', fontsize=15)
plt.legend(title='Statistics', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Add background color to enhance contrast
ax.set_facecolor('#f5f5f5')

# Show plot
plt.tight_layout()  # Adjust layout to fit all elements
plt.show()
````

### 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(0)

# Generate synthetic dataset
n = 100
data = pd.DataFrame({
    'Feature1': np.random.normal(loc=0, scale=1, size=n),
    'Feature2': np.random.normal(loc=0, scale=1, size=n),
    'Feature3': np.random.normal(loc=0, scale=1, size=n)
})

# Introduce correlation between features
data['Feature2'] = data['Feature1'] * 0.8 + data['Feature2'] * 0.2
data['Feature3'] = data['Feature1'] * 0.5 + data['Feature3'] * 0.5

# Compute the correlation matrix
corr_matrix = data.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(10, 8))
ax = sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', 
                 vmin=-1, vmax=1, center=0, linewidths=0.5, linecolor='black')

# Customize plot aesthetics
plt.title('Correlation Matrix Heatmap', fontsize=18, fontweight='bold')
plt.xlabel('Features', fontsize=15)
plt.ylabel('Features', fontsize=15)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.yticks(rotation=0)   # Rotate y-axis labels for better readability

# Show plot
plt.tight_layout()  # Adjust layout to fit all elements
plt.show()
````

````python
import numpy as np
import pandas as pd
import plotly.graph_objs as go
import plotly.io as pio

# Prepare the data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create the 3D surface plot
surface = go.Surface(
    z=z,
    x=x,
    y=y,
    colorscale='Viridis',  # Use the Viridis color scale for better visual contrast
    contours={"z": {"show": True, "start": -1, "end": 1, "size": 0.2, "color":"white"}}
)

# Customize the layout
layout = go.Layout(
    title='3D Surface Plot of z = sin(sqrt(x^2 + y^2))',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z',
        camera=dict(eye=dict(x=1.5, y=1.5, z=1.5))  # Set a default 3D camera view
    ),
    width=800,
    height=600
)

# Create the figure
fig = go.Figure(data=[surface], layout=layout)

# Show plot
pio.show(fig)
````

### 4. Create a Bokeh histogram to visualize the distribution of the given data.
```python
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
```

````python
import numpy as np
from scipy.stats import norm
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool

# To display plots directly in Jupyter notebooks
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)

# Compute the histogram
hist, edges = np.histogram(data_hist, bins=30)

# Prepare the data for Bokeh
data = {
    'top': hist,
    'left': edges[:-1],
    'right': edges[1:]
}
source = ColumnDataSource(data=data)

# Create a Bokeh figure
p = figure(title="Enhanced Histogram with Gaussian Curve and Rug Plot",
           x_axis_label='Value',
           y_axis_label='Frequency',
           width=800, height=600)

# Add a quad glyph for the histogram
quad = p.quad(top='top', bottom=0, left='left', right='right', source=source,
              fill_color='navy', line_color='white', alpha=0.7)

# Add hover tooltips only to the quad (histogram bars)
hover = HoverTool(renderers=[quad], tooltips=[("Count", "@top"), ("Range", "[@left, @right]")])
p.add_tools(hover)

# Add a Gaussian distribution curve
x = np.linspace(min(data_hist), max(data_hist), 1000)
pdf = norm.pdf(x, np.mean(data_hist), np.std(data_hist))
pdf_scaled = pdf * max(hist) / max(pdf)  # Scale to match histogram height
p.line(x, pdf_scaled, line_width=2, color='orange', legend_label="Gaussian Curve")

# Add a rug plot
p.segment(x0=data_hist, y0=0, x1=data_hist, y1=0.1 * max(hist), color="black", alpha=0.3)

# Customize the plot
p.title.text_font_size = '20pt'
p.xaxis.axis_label_text_font_size = '14pt'
p.yaxis.axis_label_text_font_size = '14pt'
p.grid.grid_line_color = "white"  # Light grid lines for a cleaner look
p.legend.location = "top_left"

# Display the plot
show(p)
````

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)

````python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import LinearColorMapper, BasicTicker, ColorBar, PrintfTickFormatter, Label, ColumnDataSource
from bokeh.palettes import Viridis256, Inferno256, Plasma256, Magma256, Cividis256

# Data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Output to notebook (if using Jupyter)
output_notebook()

# Adjusted color palette for better contrast
custom_palette = ["#ffffd9", "#edf8b1", "#c7e9b4", "#7fcdbb", "#41b6c4", 
                  "#1d91c0", "#225ea8", "#253494", "#081d58"]
mapper = LinearColorMapper(palette=custom_palette, low=data_heatmap.min(), high=data_heatmap.max())

# Create the plot with increased size
p = figure(title="Heatmap",
           x_range=(x.min(), x.max()), y_range=(y.min(), y.max()),
           x_axis_label='X', y_axis_label='Y',
           tools="hover,save,pan,box_zoom,reset,wheel_zoom",
           tooltips=[("x", "@x{0.2f}"), ("y", "@y{0.2f}"), ("value", "@value{0.2f}")],
           width=700, height=500)

# Create a ColumnDataSource for the heatmap data
source = ColumnDataSource(data=dict(
    x=xx.flatten(),
    y=yy.flatten(),
    value=data_heatmap.flatten()
))

# Draw the heatmap using p.rect
p.rect('x', 'y', width=(x[1]-x[0]), height=(y[1]-y[0]), source=source,
       fill_color={'field': 'value', 'transform': mapper},
       line_color=None)  # No outlines for the rectangles

# Add the color bar
color_bar = ColorBar(color_mapper=mapper, major_label_text_font_size="8pt",
                     ticker=BasicTicker(desired_num_ticks=10),
                     formatter=PrintfTickFormatter(format="%0.2f"),
                     label_standoff=6, border_line_color=None, location=(0,0))
p.add_layout(color_bar, 'right')

# Add annotations 
high_value_indices = np.where(data_heatmap > 0.8) 
labels_source = ColumnDataSource(data=dict(
    x=xx[high_value_indices].flatten().astype(float),
    y=yy[high_value_indices].flatten().astype(float),
    text=[f"{val:.2f}" for val in data_heatmap[high_value_indices]]
))
labels = LabelSet(x='x', y='y', text='text', source=labels_source,
                  text_color='white', text_font_size='10pt', 
                  background_fill_color='black', background_fill_alpha=0.7)
p.add_layout(labels)

# Add grid lines
p.xgrid.grid_line_color = 'gray'
p.xgrid.grid_line_alpha = 0.5
p.ygrid.grid_line_color = 'gray'
p.ygrid.grid_line_alpha = 0.5

# Show the plot
show(p)
````
