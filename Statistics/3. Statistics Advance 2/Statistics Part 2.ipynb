## **1. Hypothesis Testing in Statistics**  
Hypothesis testing is a method to determine if there is enough statistical evidence to support a claim about a population parameter. It involves:  
- **Null Hypothesis ($(H_0$))**: Assumes no effect/difference.  
- **Alternative Hypothesis ($(H_1$))**: Represents the researcher’s claim.  
- **Test Statistic**: Computed from sample data (e.g., $(Z$), $(T$)).  
- **Decision Rule**: Compare the test statistic to a critical value or use a **P-value**.  

**Example**:  
Testing if a new drug lowers blood pressure:  
- $(H_0$): Drug has no effect ($(\mu = \mu_0$)).  
- $(H_1$): Drug reduces blood pressure ($(\mu < \mu_0$)).  

---

## **2. Null Hypothesis vs. Alternative Hypothesis**  
- **Null Hypothesis ($(H_0$))**: Default assumption (e.g., $(\mu = 50$)).  
- **Alternative Hypothesis ($(H_1$))**: Contradicts $(H_0$) (e.g., $(\mu \neq 50$) or $(\mu > 50$)).  

**Key Difference**:  
- $(H_0$) is tested for rejection; $(H_1$) is the claim to validate.  

---

## **3. Significance Level ($(\alpha$))**  
- The probability of rejecting $(H_0$) when it is true (**Type I error**).  
- Common values: $(\alpha = 0.05$) or $(\alpha = 0.01$).  
- **Importance**: Controls the risk of false positives.  

---

## **4. P-value**  
- The probability of observing the sample data (or more extreme results) **assuming $(H_0$) is true**.  
- **Formula**:  
  $$[
  P\text{-value} = P(\text{Test Statistic} \geq \text{Observed Value} \mid H_0)
  $$]  
**Example**:  
If $(P = 0.03$) and $(\alpha = 0.05$), reject $(H_0$).  

---

## **5. Interpreting the P-value**  
- **Low $(P$)-value** ($(P < \alpha$)): Strong evidence against $(H_0$).  
- **High $(P$)-value** ($(P \geq \alpha$)): Insufficient evidence to reject $(H_0$).  

---

## **6. Type I and Type II Errors**  
- **Type I Error**: Rejecting $(H_0$) when it is true ($(\alpha$)).  
  - Example: Concluding a drug works when it does not.  
- **Type II Error ($(\beta$))**: Failing to reject $(H_0$) when it is false.  
  - Example: Failing to detect a drug’s effect.  

---

## **7. One-Tailed vs. Two-Tailed Tests**  
- **One-Tailed**: Tests for an effect in **one direction** (e.g., $(\mu > 50$)).  
  - Critical region in one tail of the distribution.  
- **Two-Tailed**: Tests for effects in **both directions** (e.g., $(\mu \neq 50$)).  
  - Critical regions in both tails.  

---

## **8. Z-Test**  
- A hypothesis test using the **standard normal distribution** ($(Z$)-distribution).  
- **Use Case**:  
  - Population variance ($(\sigma^2$)) is known.  
  - Large sample size ($(n \geq 30$)).  

**Formula**:  
$$[
Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
$$]  
Where:  
- $(\bar{X}$) = Sample mean  
- $(\mu$) = Population mean  
- $(\sigma$) = Population standard deviation  
- $(n$) = Sample size  

---

## **9. Z-Score Calculation**  
- Measures how many standard errors the sample mean deviates from the population mean.  
$$[
Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
$$]  
**Interpretation**:  
- $(Z = 1.96$) → Sample mean is 1.96 standard errors above $(\mu$).  

---

## **10. T-Distribution**  
- A distribution with **heavier tails** than the normal distribution.  
- **Use Case**:  
  - Population variance unknown.  
  - Small sample size ($(n < 30$)).  
- **Degrees of Freedom**: $(df = n - 1$).  

---

## **11. Z-Test vs. T-Test**  
| **Z-Test** | **T-Test** |  
|------------|------------|  
| Uses $(\sigma$) (known variance). | Uses $(s$) (sample variance). |  
| Suitable for large samples. | Suitable for small samples. |  

---

## **12. T-Test**  
- Compares means using the **T-distribution**.  
- **Types**:  
  1. **One-Sample**: Compare sample mean to a known $(\mu$).  
  2. **Independent**: Compare two independent groups.  
  3. **Paired**: Compare paired observations (e.g., pre-test vs. post-test).  

**Formula (One-Sample)**:  
$$[
t = \frac{\bar{X} - \mu}{s / \sqrt{n}}
$$]  
Where $(s$) = sample standard deviation.  

---

## **13. Relationship Between Z-Test and T-Test**  
- Both compare means, but the T-test accounts for uncertainty in estimating $(\sigma$).  
- As $(n \to \infty$), the T-distribution converges to the Z-distribution.  

---

## **14. Confidence Interval (CI)**  
- A range of values that likely contains the population parameter.  
- **Formula for Mean**:  
$$[
\text{CI} = \bar{X} \pm Z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}
$$]  
**Example**: 95% CI = $([45, 55]$) → 95% confidence the true mean lies in this interval.  

---

## **15. Margin of Error**  
- Half the width of the confidence interval.  
$$[
\text{Margin of Error} = Z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}
$$]  
- **Impact**: Larger $(n$) reduces the margin of error.  

---

## **16. Bayes’ Theorem**  
$$[
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
$$]  
- **Application**: Updates prior beliefs ($(P(A)$)) with data ($(P(B|A)$)).  
- **Example**: Spam detection (updating spam probability based on keywords).  

---

## **17. Chi-Square Distribution**  
- A right-skewed distribution for **categorical data analysis**.  
- **Use Cases**:  
  - Goodness-of-fit tests.  
  - Tests of independence.  

---

## **18. Chi-Square Goodness-of-Fit Test**  
- Tests if observed data matches an expected distribution.  
- **Formula**:  
$$[
\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
$$]  
Where:  
- $(O_i$) = Observed frequency  
- $(E_i$) = Expected frequency  

**Example**: Testing if a die is fair ($(E_i = \frac{n}{6}$)).  

---

## **19. F-Distribution**  
- The distribution of the ratio of two chi-square variables.  
- **Use Cases**:  
  - ANOVA (comparing group variances).  
  - Testing equality of variances.  

---

## **20. ANOVA Test**  
- **Analysis of Variance (ANOVA)** tests differences between group means.  
- **Assumptions**:  
  1. Normality.  
  2. Homogeneity of variances.  
  3. Independence.  

**Formula (F-statistic)**:  
$$[
F = \frac{\text{Between-group variance}}{\text{Within-group variance}}
$$]  

---

## **21. Types of ANOVA**  
1. **One-Way ANOVA**: Tests one factor (e.g., effect of fertilizer on plant growth).  
2. **Two-Way ANOVA**: Tests two factors and their interaction.  
3. **MANOVA**: Tests multiple dependent variables.  

---

## **22. F-Test**  
- Compares variances using the **F-distribution**.  
- **In ANOVA**: Compares variability between groups to variability within groups.  

**Formula**:  
$$[
F = \frac{MS_{\text{between}}}{MS_{\text{within}}}
$$]  
Where $(MS$) = Mean Square.

#### **Question 1: Z-Test for Sample Mean vs. Population Mean**  
**Problem**: Compare a sample mean to a population mean (μ = 100, σ = 15).  
**Sample Data**: `[105, 110, 95, 100, 98]`  

```python
import numpy as np
from scipy.stats import norm

# Population parameters
mu = 100
sigma = 15

# Sample data
sample = np.array([105, 110, 95, 100, 98])
n = len(sample)
x_bar = np.mean(sample)

# Z-score calculation
z = (x_bar - mu) / (sigma / np.sqrt(n))

# Two-tailed P-value
p_value = 2 * (1 - norm.cdf(abs(z)))

# Output results
print(f"Z-score: {z:.2f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject H₀: Significant difference detected.")
else:
    print("Fail to reject H₀: No significant difference.")
```

**Interpretation**:  
The sample mean (101.6) is not significantly different from the population mean (100) at α = 0.05.

#### **Question 3: One-Sample Z-Test Function**  
**Problem**: Implement a reusable Z-test function.  

```python
def one_sample_z_test(sample, pop_mean, pop_std):
    n = len(sample)
    x_bar = np.mean(sample)
    z = (x_bar - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * (1 - norm.cdf(abs(z)))
    return z, p_value

# Example usage
sample = [105, 110, 95, 100, 98]
z, p = one_sample_z_test(sample, 100, 15)
print(f"Z: {z:.2f}, P: {p:.4f}")
```

#### **Question 5: Visualizing Type I and Type II Errors**  
**Problem**: Create a function to visualize Type I/II errors.  

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

def plot_errors(alpha=0.05, effect_size=5, pop_std=15, n=30):
    # H0: μ=100; H1: μ=105 (effect size=5)
    critical_value = norm.ppf(1 - alpha, loc=100, scale=pop_std/np.sqrt(n))
    
    # Distributions
    x = np.linspace(90, 115, 1000)
    y_h0 = norm.pdf(x, 100, pop_std/np.sqrt(n))
    y_h1 = norm.pdf(x, 100 + effect_size, pop_std/np.sqrt(n))
    
    # Plot
    plt.plot(x, y_h0, label='H₀ (μ=100)')
    plt.plot(x, y_h1, label='H₁ (μ=105)')
    
    # Type I Error (α)
    plt.fill_between(x, y_h0, where=(x >= critical_value), color='red', alpha=0.3, label='Type I Error (α)')
    
    # Type II Error (β)
    beta_region = x < critical_value
    plt.fill_between(x, y_h1, where=beta_region, color='blue', alpha=0.3, label='Type II Error (β)')
    
    plt.title('Type I and Type II Errors')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.legend()
    plt.show()

plot_errors()
```

#### **Question 9: How to calculate the confidence interval for a sample mean and explain its significance**  

```python
import numpy as np
import scipy.stats as stats

# Define the sample data
sample_mean = 75
sample_std = 10
sample_size = 100

# Calculate the standard error of the mean
sem = sample_std / (sample_size ** 0.5)

# Calculate the degrees of freedom
df = sample_size - 1

# Calculate the t-value for a 95% confidence level
confidence_level = 0.95
alpha = 1 - confidence_level
t_value = stats.t.ppf(1 - alpha / 2, df)

# Calculate the margin of error
margin_of_error = t_value * sem

# Calculate the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Print the confidence interval
print("95% Confidence Interval:", (lower_bound, upper_bound))
```

#### **Question 11: How to implement a Bayesian inference method using Bayes' Theorem and explain the process**  

```python
import numpy as np

def bayesian_inference(prior, likelihood, evidence):
    """
    Performs Bayesian inference.

    Args:
        prior (numpy.ndarray): Prior probabilities.
        likelihood (numpy.ndarray): Likelihood probabilities.
        evidence (float): Probability of the evidence.

    Returns:
        numpy.ndarray: Posterior probabilities.
    """

    # Calculate the joint probabilities
    joint = prior * likelihood

    # Calculate the posterior probabilities
    posterior = joint / evidence

    return posterior

# Example usage
prior = np.array([0.6, 0.4])  # Prior probabilities
likelihood = np.array([0.8, 0.3])  # Likelihood probabilities
evidence = 0.6  # Probability of the evidence

# Perform Bayesian inference
posterior = bayesian_inference(prior, likelihood, evidence)

# Print the posterior probabilities
print("Posterior probabilities:", posterior)
```

#### **Question 13: How to calculate the expected frequencies for a Chi-square test based on observed data**  

```python
import numpy as np

def chi2_expected_frequencies(observed):
    """
    Calculates the expected frequencies for a Chi-square test.

    Args:
        observed (list of lists): Observed frequencies.

    Returns:
        numpy.ndarray: Expected frequencies.
    """

    # Convert the observed frequencies to a NumPy array
    observed = np.array(observed)

    # Calculate the row and column sums
    row_sums = observed.sum(axis=1)
    col_sums = observed.sum(axis=0)

    # Calculate the total sum
    total_sum = observed.sum()

    # Calculate the expected frequencies
    expected = np.outer(row_sums, col_sums) / total_sum

    return expected

# Example usage
observed = [[25, 30], [35, 40]]

# Calculate the expected frequencies
expected = chi2_expected_frequencies(observed)

# Print the expected frequencies
print("Expected frequencies:", expected)
```

#### **Question 14: How to perform a goodness-of-fit test to compare the observed data to an expected distribution**  

```python
import scipy.stats as stats
import numpy as np

def chi_square_goodness_of_fit(observed, expected):
    """
    Performs a Chi-square goodness-of-fit test, handling potential sum mismatches.

    Args:
        observed (list or numpy.ndarray): Observed frequencies.
        expected (list or numpy.ndarray): Expected frequencies.

    Returns:
        tuple: (chi2_stat, p_value) or None if error occurs.
    """

    observed = np.array(observed)
    expected = np.array(expected)

    # Check if the sums are equal, adjust if needed
    observed_sum = np.sum(observed)
    expected_sum = np.sum(expected)

    if observed_sum != expected_sum:
        print("Warning: Sum of observed and expected frequencies do not match.")
        
        # Adjust expected frequencies to match observed total.
        scale_factor = observed_sum / expected_sum
        expected = expected * scale_factor
        print("Expected frequencies adjusted to match observed total.")

    try:
        chi2_stat, p_value = stats.chisquare(observed, expected)
        return chi2_stat, p_value
    except ValueError as e:
        print(f"Error during Chi-square test: {e}")
        return None  # Return None to indicate an error

# Example usage (with the error-causing data)
observed = [25, 30, 35, 40, 45]
expected = [20, 25, 30, 35, 40]

results = chi_square_goodness_of_fit(observed, expected)

if results:
    chi2_stat, p_value = results
    alpha = 0.05
    if p_value < alpha:
        print("The observed data does not fit the expected distribution.")
    else:
        print("The observed data fits the expected distribution.")
    print("Chi-square statistic:", chi2_stat)
    print("P-value:", p_value)
```

#### **Question 16: How to implement an F-test to compare the variances of two random samples**  

```python
import numpy as np

import scipy.stats as stats

# Define the two random samples
sample1 = [25, 30, 32, 28, 29, 35, 31, 33, 27, 30, 32, 34, 28, 29, 31, 26, 33, 27, 30, 32]
sample2 = [22, 28, 30, 25, 26, 32, 29, 31, 24, 27, 29, 33, 26, 28, 30, 23, 31, 25, 28, 30]

# Calculate the variances of the two samples
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Perform the F-test
f_stat = var1 / var2
dof1 = len(sample1) - 1
dof2 = len(sample2) - 1
p_value = 1 - stats.f.cdf(f_stat, dof1, dof2)

# Check the p-value to determine statistical significance
alpha = 0.05  # Significance level
if p_value < alpha:
    print("The variances of the two samples are significantly different.")
else:
    print("The variances of the two samples are not significantly different.")

# Print the F-statistic and p-value
print("F-statistic:", f_stat)
print("P-value:", p_value)
```

#### **Question 19: How to check the assumptions (normality, independence, and equal variance) for ANOVA**  

```python
import numpy as np
import seaborn as sns

import scipy.stats as stats
import matplotlib.pyplot as plt

# Define the data for multiple groups
group1 = [25, 30, 32, 28, 29, 35, 31, 33, 27, 30, 32, 34, 28, 29, 31, 26, 33, 27, 30, 32]
group2 = [22, 28, 30, 25, 26, 32, 29, 31, 24, 27, 29, 33, 26, 28, 30, 23, 31, 25, 28, 30]
group3 = [28, 33, 35, 30, 32, 38, 34, 36, 29, 33, 35, 37, 30, 32, 34, 28, 36, 29, 33, 35]

# Combine the groups into a single dataset
data = [group1, group2, group3]

# Check normality using Shapiro-Wilk test
for i, group in enumerate(data, start=1):
    stat, p = stats.shapiro(group)
    print(f'Group {i} - Shapiro-Wilk Test: Statistics={stat:.3f}, p={p:.3f}')
    if p > 0.05:
        print(f'Group {i} looks Gaussian (fail to reject H0)')
    else:
        print(f'Group {i} does not look Gaussian (reject H0)')

# Check independence (this is usually ensured by the study design)

# Check homogeneity of variances using Levene's test
stat, p = stats.levene(group1, group2, group3)
print(f'\nLevene’s Test for Homogeneity of Variances: Statistics={stat:.3f}, p={p:.3f}')
if p > 0.05:
    print('Variances are equal (fail to reject H0)')
else:
    print('Variances are not equal (reject H0)')

# Visualize the data
plt.figure(figsize=(10, 6))
sns.boxplot(data=[group1, group2, group3])
plt.title('Boxplot of Groups')
plt.xlabel('Group')
plt.ylabel('Values')
plt.xticks([0, 1, 2], ['Group 1', 'Group 2', 'Group 3'])
plt.show()
```

#### **Question 20: How to perform a two-way ANOVA test to study the interaction between two factors and visualize the results**  

```python
import pandas as pd
from statsmodels.formula.api import ols
import seaborn as sns

import statsmodels.api as sm
import matplotlib.pyplot as plt

# Example data
data = {
    'Factor1': ['A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'C', 'C', 'C', 'C'],
    'Factor2': ['X', 'X', 'Y', 'Y', 'X', 'X', 'Y', 'Y', 'X', 'X', 'Y', 'Y'],
    'Value': [23, 27, 24, 25, 30, 31, 29, 28, 35, 36, 34, 33]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Perform two-way ANOVA
model = ols('Value ~ C(Factor1) + C(Factor2) + C(Factor1):C(Factor2)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)

# Visualize the interaction
sns.pointplot(x='Factor1', y='Value', hue='Factor2', data=df, dodge=True, markers=['o', 's'], capsize=.1, errwidth=1, palette='colorblind')
plt.title('Interaction Plot')
plt.show() 
```

#### **Question 21: How to visualize the F-distribution and discuss its use in hypothesis testing**  

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define the degrees of freedom
dfn = 5  # numerator degrees of freedom
dfd = 20  # denominator degrees of freedom

# Simulate the F-distribution
f_samples = np.random.f(dfn, dfd, size=1000)

# Plot the histogram of the samples
plt.hist(f_samples, bins=30, density=True, alpha=0.7, label='F-distribution samples')

# Plot the PDF of the F-distribution
x = np.linspace(stats.f.ppf(0.01, dfn, dfd), stats.f.ppf(0.99, dfn, dfd), 100)
plt.plot(x, stats.f.pdf(x, dfn, dfd), 'r-', lw=2, label='F-distribution PDF')

# Set the title and labels
plt.title(f'F-distribution (dfn={dfn}, dfd={dfd})')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

#### **Question 22: How to perform a one-way ANOVA test and visualize the results with boxplots to compare group means**  

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

# Define the data for different groups
group1 = [25, 30, 32, 28, 29, 35, 31, 33, 27, 30, 32, 34, 28, 29, 31, 26, 33, 27, 30, 32]
group2 = [22, 28, 30, 25, 26, 32, 29, 31, 24, 27, 29, 33, 26, 28, 30, 23, 31, 25, 28, 30]
group3 = [28, 33, 35, 30, 32, 38, 34, 36, 29, 33, 35, 37, 30, 32, 34, 28, 36, 29, 33, 35]

# Perform the one-way ANOVA test
f_stat, p_value = stats.f_oneway(group1, group2, group3)

# Check the p-value to determine statistical significance
alpha = 0.05  # Significance level
if p_value < alpha:
    print("There is a significant difference between the means of the groups.")
else:
    print("There is no significant difference between the means of the groups.")

# Print the F-statistic and p-value
print("F-statistic:", f_stat)
print("P-value:", p_value)

# Visualize the results using box plots
plt.boxplot([group1, group2, group3], labels=['Group 1', 'Group 2', 'Group 3'])
plt.title('Comparison of Group Means')
plt.xlabel('Group')
plt.ylabel('Values')
plt.show()
```

#### **Question 23: How to simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means**  

```python
import numpy as np
import scipy.stats as stats

# Simulate random data from a normal distribution
np.random.seed(0)  # For reproducibility
data = np.random.normal(loc=0, scale=1, size=30)  # Simulate data from a standard normal distribution

# Known population parameters
population_mean = 0
population_std = 1

# Perform one-sample z-test
z_stat, z_p_value = stats.ttest_1samp(data, population_mean)

# Perform one-sample t-test
t_stat, t_p_value = stats.ttest_1samp(data, population_mean)

# Compare the results
print("Z-test results:")
print("Z-statistic:", z_stat)
print("P-value:", z_p_value)

print("\nT-test results:")
print("T-statistic:", t_stat)
print("P-value:", t_p_value)

# Compare the p-values
if z_p_value < 0.05:
    print("\nZ-test: Reject null hypothesis")
else:
    print("\nZ-test: Fail to reject null hypothesis")

if t_p_value < 0.05:
    print("T-test: Reject null hypothesis")
else:
    print("T-test: Fail to reject null hypothesis")
```

#### **Question 24: How to perform a hypothesis test for population variance using a Chi-square distribution and interpret the results**  

```python
import numpy as np
import scipy.stats as stats

# Define the sample data
sample_data = [25, 30, 32, 28, 29, 35, 31, 33, 27, 30, 32, 34, 28, 29, 31, 26, 33, 27, 30, 32]

# Calculate the sample variance
sample_variance = np.var(sample_data, ddof=1)  # ddof=1 for sample variance

# Define the hypothesized population variance
hypothesized_variance = 10

# Calculate the test statistic
n = len(sample_data)
chi2_stat = (n - 1) * sample_variance / hypothesized_variance

# Calculate the p-value
p_value = stats.chi2.sf(chi2_stat, n - 1)

# Check the p-value to determine statistical significance
alpha = 0.05  # Significance level
if p_value < alpha:
    print("The population variance is significantly different from the hypothesized variance.")
else:
    print("The population variance is not significantly different from the hypothesized variance.")

# Print the test statistic and p-value
print("Chi-square statistic:", chi2_stat)
print("P-value:", p_value)
```

#### **Question 25: Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.**  

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def z_test_proportions(count1, nobs1, count2, nobs2, alternative='two-sided'):
    """Performs a Z-test for comparing proportions between two groups."""
    p1 = count1 / nobs1
    p2 = count2 / nobs2
    p_pooled = (count1 + count2) / (nobs1 + nobs2)
    se_pooled = np.sqrt(p_pooled * (1 - p_pooled) * (1/nobs1 + 1/nobs2))
    z_stat = (p1 - p2) / se_pooled

    if alternative == 'two-sided':
        p_value = 2 * stats.norm.sf(np.abs(z_stat))
    elif alternative == 'greater':
        p_value = stats.norm.sf(z_stat)
    elif alternative == 'less':
        p_value = stats.norm.cdf(z_stat)
    else:
        raise ValueError("Alternative must be one of: 'two-sided', 'greater', 'less'")

    return z_stat, p_value

# Example usage:
count1, nobs1 = 50, 200
count2, nobs2 = 70, 250

z_stat, p_value = z_test_proportions(count1, nobs1, count2, nobs2)
print(f"Z-statistic: {z_stat:.4f}, P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject null hypothesis: proportions are different.")
else:
    print("Fail to reject null hypothesis: proportions are not significantly different.")
```

#### **Question 26: Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.**  

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def f_test_variances(data1, data2):
    """Performs an F-test for comparing variances of two datasets."""
    var1 = np.var(data1, ddof=1)
    var2 = np.var(data2, ddof=1)
    f_stat = var1 / var2
    dfn = len(data1) - 1
    dfd = len(data2) - 1
    p_value = stats.f.sf(f_stat, dfn, dfd)

    return f_stat, p_value

# Example Usage:
data1 = [25, 30, 32, 28, 29, 35, 31, 33, 27, 30]
data2 = [22, 28, 30, 25, 26, 32, 29, 31, 24, 27]

f_stat, p_value = f_test_variances(data1, data2)
print(f"F-statistic: {f_stat:.4f}, P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject null hypothesis: variances are different.")
else:
    print("Fail to reject null hypothesis: variances are not significantly different.")

# Visualize the F-distribution
dfn = len(data1) -1
dfd = len(data2) -1
x = np.linspace(stats.f.ppf(0.01, dfn, dfd), stats.f.ppf(0.99, dfn, dfd), 100)
plt.plot(x, stats.f.pdf(x, dfn, dfd), 'r-', lw=2, label='F-distribution PDF')
plt.axvline(f_stat, color='b', linestyle='--', label=f'F-statistic = {f_stat:.2f}')
plt.title(f'F-distribution (dfn={dfn}, dfd={dfd})')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

#### **Question 27: Perform a Chi-square test for goodness of fit with simulated data and analyze the results.**  

```python
import numpy as np
import scipy.stats as stats

def chi2_goodness_of_fit(observed, expected):
    """Performs a Chi-square goodness-of-fit test, adjusting for sum differences."""

    observed = np.array(observed)
    expected = np.array(expected)

    # Check and adjust for sum differences
    observed_sum = np.sum(observed)
    expected_sum = np.sum(expected)

    if observed_sum != expected_sum:
        print("Warning: Sum of observed and expected frequencies do not match.")
        scale_factor = observed_sum / expected_sum
        expected = expected * scale_factor
        print("Expected frequencies adjusted to match observed total.")

    try:
        chi2_stat, p_value = stats.chisquare(observed, expected)
        return chi2_stat, p_value
    except ValueError as e:
        print(f"Error during Chi-square test: {e}")
        return None, None

# Example Usage:
observed = [25, 30, 35, 40, 45]
expected = [22, 28, 32, 38, 48] # changed expected to avoid the same sums.

chi2_stat, p_value = chi2_goodness_of_fit(observed, expected)

if chi2_stat is not None and p_value is not None:
    print(f"Chi-square statistic: {chi2_stat:.4f}, P-value: {p_value:.4f}")

    if p_value < 0.05:
        print("Reject null hypothesis: observed data does not fit expected distribution.")
    else:
        print("Fail to reject null hypothesis: observed data fits expected distribution.")
else:
    print("Chi-square test failed.")
```
